# Joining DataFrames

```python
teaching = [{"name":"ahmet bulut", "code":340},
            {"name":"baris arslan", "code":372}]
courses = [{"name":"computer systems", "code":340, "room":"G218"}, 
           {"name":"advanced algorithms", "code":372, "room":"G211"}]

dfTeaching = sqlContext.createDataFrame(sc.parallelize(teaching))
dfCourses = sqlContext.createDataFrame(sc.parallelize(courses))
```

```python
dfTeaching.join(dfCourses, dfTeaching.code == dfCourses.code).show()
```

```python
dx = dfTeaching.join(dfCourses, dfTeaching.code == dfCourses.code).select(dfTeaching.name.alias("teacher_name"), dfCourses.name.alias("course_name")).show()
```
